MLP

```python
# Import libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error
```

```python
# Assuming 'news' is a Pandas DataFrame with 'text' and 'label' columns
# For demonstration purposes, let's assume 'news' is already defined
text = news["text"]
label = news["label"]
```

```python
# Feature engineering: TF-IDF Vectorization
vectorizer = TfidfVectorizer(max_features=10000)  # Adjust max_features as needed
features = vectorizer.fit_transform(text)
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=42)
```

```python
# Model building and training
model = MLPRegressor(solver='adam', hidden_layer_sizes=(100, 50), activation='relu', random_state=42)
model.fit(X_train, y_train)
```

```python
# Evaluation
# Predict on the test set
y_pred = model.predict(X_test)
```

```python
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE) on test set: {mae}")
```

```python
# Prediction function for new articles
def predict_mlp(article_string):
    # Preprocess the article string (e.g., tokenization, cleaning)
    article_vector = vectorizer.transform([article_string])
    prediction = model.predict(article_vector)[0]
    if prediction > 0.5:  # Adjust threshold based on model output
        return "Real News"
    else:
        return "Fake News"
```

```python
new_article = str(input())
prediction = predict_mlp(new_article)
print(f"Predicted label for '{new_article}': \n{prediction}")
```
