```python
news = news.drop(['title', 'subject', 'date'], axis=1)
```

```python
news = news.sample(frac=1) #Reshuffling
```

CNN

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Assuming 'news' is a Pandas DataFrame with 'text' and 'label' columns
# For demonstration purposes, let's assume 'news' is already defined
text = news["text"]
label = news["label"]

# Encode labels
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(label)

# Tokenization and padding
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(text)
sequences = tokenizer.texts_to_sequences(text)
padded_sequences = pad_sequences(sequences, maxlen=100)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2, random_state=42)

# Model building and training (CNN)
model = Sequential()
model.add(Embedding(10000, 128))
model.add(Conv1D(64, 5, activation='relu'))
model.add(GlobalMaxPooling1D())
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))

# Evaluation
# Predict on the test set
y_pred_proba = model.predict(X_test)
y_pred = (y_pred_proba > 0.5).astype(int)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE) on test set: {mae}")

# Prediction function for new articles
def predict_cnn(article_string):
    # Preprocess the article string (e.g., tokenization, padding)
    sequence = tokenizer.texts_to_sequences([article_string])
    padded_sequence = pad_sequences(sequence, maxlen=100)
    prediction = model.predict(padded_sequence)[0]
    if prediction > 0.5:
        return "Real News"
    else:
        return "Fake News"
```

```python
# Example usage
new_article = str(input())
prediction = predict_cnn(new_article)
print(f"Predicted label for '{new_article}':\n {prediction}")
```
